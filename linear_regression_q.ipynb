1. All data


```python
import pandas as pd

data = pd.read_csv('_Cities.csv')
print(data)
```

2. All Cities and Countries

```python
print(data[['city', 'country']])
```

3. All data in Italy

```python
print(data[(data.country == 'Italy')])
```

4. All Cities in Spain

```python
spain_data = data[data['country'] == 'Spain']
print(spain_data)
```

5. First 10 data with a latitude of 40 or above

```python
data_filtered = data[data['latitude'] > 40]
print(data_filtered.head(10))
```

6. List 20 cities and countries with countries that has spaces in them

```python
res = data['country'].str.endswith('')
city_country = data[['city', 'country']][res].head(20)
print(city_country)
```
